```python
# imports

import pandas as pd
import numpy as np
```

```python
#read csv files in

df2019 = pd.read_csv("NBA_data/NBA_Season_avgs_2019_20.csv")
df2018 = pd.read_csv("NBA_data/NBA_Season_avgs_2018_19.csv")
df2017 = pd.read_csv("NBA_data/NBA_Season_avgs_2017_18.csv")
df2016 = pd.read_csv("NBA_data/NBA_Season_avgs_2016_17.csv")
df2015 = pd.read_csv("NBA_data/NBA_Season_avgs_2015_16.csv")
df2014 = pd.read_csv("NBA_data/NBA_Season_avgs_2014_15.csv")
```

```python
# add year columns to each

df2019['Year'] = "2019"
df2018['Year'] = "2018"
df2017['Year'] = "2017"
df2016['Year'] = "2016"
df2015['Year'] = "2015"
df2014['Year'] = "2014"
```

```python
df2019
```

```python
# Rk -- Rank
# G -- Games
# MP -- Minutes Played
# FG -- Field Goals
# FGA -- Field Goal Attempts
# FG% -- Field Goal Percentage
# 3P -- 3-Point Field Goals
# 3PA -- 3-Point Field Goal Attempts
# 3P% -- 3-Point Field Goal Percentage
# 2P -- 2-Point Field Goals
# 2PA -- 2-point Field Goal Attempts
# 2P% -- 2-Point Field Goal Percentage
# FT -- Free Throws
# FTA -- Free Throw Attempts
# FT% -- Free Throw Percentage
# ORB -- Offensive Rebounds
# DRB -- Defensive Rebounds
# TRB -- Total Rebounds
# AST -- Assists
# STL -- Steals
# BLK -- Blocks
```

```python
# combine files
train_df = pd.concat([df2015, df2014], axis=0)
train_df["playoffs_y_n"] = train_df["Team"].map(lambda x: 1 if "*" in x else 0)
train_df["Team"] = train_df["Team"].str.replace("*", "")
train_df = train_df.dropna()
train_df = train_df.reset_index()
train_df = train_df.drop(columns=["G","W","L","PF","PTS","TOV","Rk","Year","Team","MP","FG","FGA","3P","3PA","2P","2PA","FT","FTA","DRB","TRB","STL","BLK"])
```

```python
test_df = pd.concat([df2017, df2016], axis=0)
test_df["playoffs_y_n"] = test_df["Team"].map(lambda x: 1 if "*" in x else 0)
test_df["Team"] = test_df["Team"].str.replace("*", "")
test_df = test_df.dropna()
test_df = test_df.reset_index()
test_df = test_df.drop(columns=["G","W","L","PF","PTS","TOV","Rk","Year","Team","MP","FG","FGA","3P","3PA","2P","2PA","FT","FTA","DRB","TRB","STL","BLK"])
```

```python
prediction_df = df2018
prediction_df["playoffs_y_n"] = prediction_df["Team"].map(lambda x: 1 if "*" in x else 0)
prediction_df["Team"] = prediction_df["Team"].str.replace("*", "")
prediction_df = prediction_df.dropna()
prediction_df = prediction_df.reset_index()
prediction_df = prediction_df.drop(columns=["G","W","L","PF","PTS","TOV","Rk","Year","Team","MP","FG","FGA","3P","3PA","2P","2PA","FT","FTA","DRB","TRB","STL","BLK","playoffs_y_n"])
```

```python
prediction_df
```

```python
# train data = 2014, 2015, 2016, 2017
# test data = 2018, 2019
```

```python
y_test = test_df["playoffs_y_n"]
X_test = test_df.drop(columns=["playoffs_y_n"])

y_train = train_df["playoffs_y_n"]
X_train = train_df.drop(columns=["playoffs_y_n"])
```

```python
from sklearn.preprocessing import MinMaxScaler
X_scaler = MinMaxScaler().fit(X_train)

X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
print(f"Training Data Score: {model.score(X_train, y_train)}")
print(f"Testing Data Score: {model.score(X_test, y_test)}")
```

```python
from sklearn.model_selection import GridSearchCV
param_grid = {'C': [1, 5, 10],
              'penalty': ["0.01", "0.001","0.0001"]}
grid = GridSearchCV(model, param_grid, verbose=3)
```

```python
grid.fit(X_train, y_train)
```

```python
print(grid.best_params_)
print(grid.best_score_)
```
