```python
# imports

import pandas as pd
import numpy as np
```

```python
#read csv files in

df2019 = pd.read_csv("NBA_data/NBA_Season_avgs_2019_20.csv")
df2018 = pd.read_csv("NBA_data/NBA_Season_avgs_2018_19.csv")
df2017 = pd.read_csv("NBA_data/NBA_Season_avgs_2017_18.csv")
df2016 = pd.read_csv("NBA_data/NBA_Season_avgs_2016_17.csv")
df2015 = pd.read_csv("NBA_data/NBA_Season_avgs_2015_16.csv")
df2014 = pd.read_csv("NBA_data/NBA_Season_avgs_2014_15.csv")
```

```python
# add year columns to each

df2019['Year'] = "2019"
df2018['Year'] = "2018"
df2017['Year'] = "2017"
df2016['Year'] = "2016"
df2015['Year'] = "2015"
df2014['Year'] = "2014"
```

```python
# combine files
final_df = pd.concat([df2019, df2018, df2017, df2016, df2015, df2014], axis=0)
```

```python
final_df
```

```python
final_df = final_df.dropna()
final_df["playoffs_y_n"] = np.where(final_df['Rk'] <= 16, 1, 0)
final_df
```

```python
final_df_2018 = final_df.loc[final_df["Year"] == '2018']
final_df_2017 = final_df.loc[final_df["Year"] == '2017']
final_df_2016 = final_df.loc[final_df["Year"] == '2016']
final_df_2015 = final_df.loc[final_df["Year"] == '2015']
```

# Create a Train Test Split

Use `koi_disposition` for the y values

```python
from sklearn.model_selection import train_test_split

# y = df2018["playoffs_y_n"]
# X = df2018.drop(columns=["Team", 'Rk', 'playoffs_y_n', 'Year'])
```

```python
# X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1, stratify=y)

X_train = final_df_2017.drop(columns=["Team", 'Rk', 'playoffs_y_n', 'Year'])
X_test = final_df_2018.drop(columns=["Team", 'Rk', 'playoffs_y_n', 'Year'])
y_train = final_df_2017["playoffs_y_n"]
y_test = final_df_2018["playoffs_y_n"]
```

```python
X_train
```

# Pre-processing

Scale the data using the MinMaxScaler

```python
from sklearn.preprocessing import MinMaxScaler
X_scaler = MinMaxScaler().fit(X_train)

X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

# Train the Model

```python
# Train the Model# Support vector machine linear classifier
from sklearn.svm import SVC 
model2 = SVC(kernel='linear')
model2.fit(X_train_scaled, y_train)
```

```python
print(f"Training Data Score: {model2.score(X_train_scaled, y_train)}")
print(f"Testing Data Score: {model2.score(X_test_scaled, y_test)}")
```

# Hyperparameter Tuning

Use `GridSearchCV` to tune model's parameters

```python
from sklearn.model_selection import GridSearchCV
param_grid = {'C': [1, 5, 10],
              'gamma': [0.0001, 0.001, 0.01]}
grid = GridSearchCV(model2, param_grid, verbose=3)
```

```python
grid.fit(X_train_scaled, y_train)
```

```python
print(grid.best_params_)
print(grid.best_score_)
```
